### **Załadowanie bibliotek**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Wczytanie i podgląd danych**

```python
data = pd.read_csv('dane_z_badan_dzwignicy.csv')
print('Shape: ', data.shape)
data.head(10)
```

```python
data.describe()
```

### **Sprawdzenie wartości brakujących**

```python
data.isna().sum()
```

```python
((data['Maksymalne, chwilowe zużycie ON [l/h]'].isna()) & (data['Maksymalne, chwilowe zużycie energii elektrycznej [kW]'].isna())).sum()
```

## **Inżynieria cech i eksploracyjna analiza danych**
### **Usunięcie brakujących wartości z kolumny docelowej**

```python
data.dropna(subset=['Koszt dzienny [PLN]'], inplace=True)
print('Shape: ', data.shape)
data.isna().sum()
```

### **Wypełnienie brakujących wartości i usunięcie kolumn**
Brakujące wartości w kolumnach dotyczących zurzycia paiwa/energii wypełniamy zerami, natomiast ceny wypełniamy medianą. Usuwamy kolumnę z ceną energii elektrycznej, gdyż jest ona stała oraz jedną kolumn, która przechowuje informacje o prędkości wiatru.

```python
data['Maksymalne, chwilowe zużycie ON [l/h]'] = data['Maksymalne, chwilowe zużycie ON [l/h]'].fillna(0)
data['Dzienne zużycie ON [l/8h]'] = data['Dzienne zużycie ON [l/8h]'].fillna(0)
data['Cena hurtowa ON 1000l [PLN]'] = data['Cena hurtowa ON 1000l [PLN]'].fillna(data['Cena hurtowa ON 1000l [PLN]'].median())
data['Maksymalne, chwilowe zużycie energii elektrycznej [kW]'] = data['Maksymalne, chwilowe zużycie energii elektrycznej [kW]'].fillna(0)
data['Dzienne zużycie energii elektrycznej [kW/8h]'] = data['Dzienne zużycie energii elektrycznej [kW/8h]'].fillna(0)
data.drop(columns=['Cena energii elektrycznej [kWh]', 'Prędkość wiatru [km/h]', 'Nr. Pomiaru'], inplace=True)
data.head(10)
```

### **Wizualzacja kosztu dziennego z podziałem na rodzaj napędu**

```python
data['Diesel'] = (data['Dzienne zużycie ON [l/8h]'] != 0).astype(int)
data['Data'] = pd.to_datetime(data['Data'])
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15, 7))
#fig.subplots_adjust(wspace=1)
ax[0].plot(data[data['Diesel'] == 1]['Data'], data[data['Diesel'] == 1]['Koszt dzienny [PLN]'])
ax[1].plot(data[data['Diesel'] == 0]['Data'], data[data['Diesel'] == 0]['Koszt dzienny [PLN]'])
ax[0].set_title('Diesel', fontweight='bold'); ax[0].set_ylabel('Koszt dzienny [PLN]')
ax[1].set_title('Elektryczny', fontweight='bold'); ax[1].set_ylabel('Koszt dzienny [PLN]')
plt.tight_layout()
plt.show()
```

### **Dodanie informacji o dniu tygodnia, miesiącu i roku**
W celu sprawdzenia okresowości ceny dodamy informacje o dniu tygodnia, miesiącu i roku.

```python
data['Dzień tygodnia'] = data['Data'].dt.weekday
data['Dzień tygodnia'] += 1
data['Miesiąc'] = data['Data'].dt.month
data['Rok'] = data['Data'].dt.year
data.head()
```

### **Sprawdzenie okresowości cen w skali tygodnia oraz roku**

```python
fig, ax = plt.subplots(2, 3, figsize=(12, 8))
ax = ax.flatten()
sns.boxplot(x='Dzień tygodnia', y='Koszt dzienny [PLN]', data=data[data['Diesel'] == 1], ax=ax[0])
sns.boxplot(x='Miesiąc', y='Koszt dzienny [PLN]', data=data[data['Diesel'] == 1], ax=ax[1])
sns.boxplot(x='Rok', y='Koszt dzienny [PLN]', data=data[data['Diesel'] == 1], ax=ax[2])
sns.boxplot(x='Dzień tygodnia', y='Koszt dzienny [PLN]', data=data[data['Diesel'] == 0], ax=ax[3])
sns.boxplot(x='Miesiąc', y='Koszt dzienny [PLN]', data=data[data['Diesel'] == 0], ax=ax[4])
sns.boxplot(x='Rok', y='Koszt dzienny [PLN]', data=data[data['Diesel'] == 0], ax=ax[5])
ax[1].set_title('Diesel'); ax[4].set_title('Elektryczny')
fig.suptitle('Koszt dzienny w danych miesiącach i dniach tygodnia')
plt.tight_layout()
plt.show()
```

### **Usunięcie informacji o roku**

```python
data.drop(columns=['Rok'], inplace=True)
```
